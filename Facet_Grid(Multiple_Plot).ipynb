### Facet Grid (Multiple Plot) ----------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d=sns.load_dataset("tips")
d
```

```python
f=sns.FacetGrid(d,col="sex",hue="day")

f.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
f=sns.FacetGrid(d,col="day",hue="sex")

f.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
f=sns.FacetGrid(d,col="day",hue="sex")

f.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
f=sns.FacetGrid(d,col="day",hue="sex",palette="cool_r")

f.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```
