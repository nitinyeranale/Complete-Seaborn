### Strip Plot--------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d=sns.load_dataset("tips")
d
```

```python
sns.stripplot(x="day",y="total_bill",data=d,hue="sex",palette="crest_r")
plt.legend(loc="upper left")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=d,hue="sex",palette="crest_r",jitter=1,size=5,alpha=0.5)
plt.legend(loc="upper left")
plt.show()
```

```python
sns.stripplot(x=d["total_bill"])
plt.show()
```
