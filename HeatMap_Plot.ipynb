### HeatMap Plot ----------------------------

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var=np.linspace(1,10,30).reshape(6,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
D=sns.load_dataset("anagrams")
x=D.drop(columns=["attnr"],axis=1).head(10)
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,vmin=0,vmax=10,cmap="PuOr",annot=True)
plt.show()
```

##### fmt Parameter -------

```python
m=np.linspace(1,10,10).reshape(2,5)
m
```

```python
n=np.array([["A","A1","A2","A3","A4"],
           ["B","B1","B2","B3","B4"]])
n
```

```python
sns.heatmap(m,cmap="PuOr",annot=n,fmt="s")
plt.show()
```

###### Another method ----

```python
m=np.linspace(1,5,10).reshape(2,5)
m
```

```python
n=np.array([["A","A1","A2","A3","A4"],
           ["B","B1","B2","B3","B4"]])
n
```

```python
sns.heatmap(m,cmap="PuOr",annot=True,fmt=".5f",linewidth=3,linecolor="g",annot_kws={"fontsize":10,"color":"y"})
plt.show()
```
