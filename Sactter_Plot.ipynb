### Scatter Plot --------------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
D=sns.load_dataset("penguins").head(20)
D
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=D,hue="sex",style="sex")
plt.legend(loc="lower right",title="Legend")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=D,hue="sex",size="sex",sizes=(50,40),palette="pink")
plt.legend(loc="lower right",title="Legend")
plt.show()
```
