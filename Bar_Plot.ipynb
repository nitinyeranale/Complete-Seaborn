### Bar Plot --------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
D=sns.load_dataset("penguins")
D
```

```python
sns.barplot(x="island",y="bill_length_mm",data=D)
# sns.barplot(x=D.island,y=D.bill_length_mm)

plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=D,hue="sex")
# sns.barplot(x=D.island,y=D.bill_length_mm,hue="sex")
plt.show()
```

```python
# ord=["Dream","Torgersen","Biscoe"]
# sns.barplot(x="island",y="bill_length_mm",data=D,hue="sex",order=ord)
sns.barplot(x="island",y="bill_length_mm",data=D,hue="sex",order=["Dream","Torgersen","Biscoe"],hue_order=["Female","Male"])

plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=D,hue="sex",order=["Dream","Torgersen","Biscoe"],hue_order=["Female","Male"],errorbar=('ci', 0))
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=D,hue="sex",saturation=1,errcolor="y",errwidth=10,dodge=False)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=D,hue="sex")
plt.show()
```
