### Histogram Plot ------------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
D=sns.load_dataset("penguins")
D
```

```python
sns.displot(D["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True)
plt.show()
```

```python
sns.displot(D["flipper_length_mm"],bins=[170,180,190,200,210,220,230],kde=True,rug=True,element="step")
plt.show()
```

```python
sns.displot(D,x="body_mass_g",hue="sex")
plt.show()
```
