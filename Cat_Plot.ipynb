### Cat Plot -----------------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d=sns.load_dataset("tips")
d
```

```python
sns.catplot(x="size",y="tip",data=d,kind="point",hue="sex") # Kind ='strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'.
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=d,hue="sex",kind="box")     # Kind ='strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'.
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=d,hue="sex",kind="bar",palette="CMRmap",height=3) # Kind ='strip','swarm','box','boxen','violin','bar','count',point'.
plt.show()
```

```python
sns.catplot(x="day",y="size",data=d,kind="boxen",hue="sex")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=d,kind="point",hue="sex",col="sex") # Kind ='strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'.
plt.show()
```
