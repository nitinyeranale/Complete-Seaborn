### how to Styling a Plot ---------------------------

#### -> figure Style
#### -> removing axis spines
#### -> scale and context

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d=sns.load_dataset("tips")
d
```

```python
sns.boxplot(=d["total_bill"])
plt.show()
```

#### Figure Style ----------

```python
sns.set_style("darkgrid")           # Seaborn grid style  -----------------------
# sns.set(style="dark")

sns.barplot(x="day",y="total_bill",data=d,hue="day", palette="CMRmap")
# plt.grid()                   # Matplotlib grid
plt.show()
```

#### Remove Axis -----------------

```python
sns.set_style("white")           # Seaborn grid
# sns.set(style="dark")

sns.barplot(x="day",y="total_bill",data=d,hue="day", palette="CMRmap")
# plt.grid()                   # Matplotlib grid


sns.despine()             # sns remove axis from plot ----------------------------------

plt.show()
```

#### Figure Scale(size) ----------------

```python
sns.set_style("white")           # Seaborn grid
# sns.set(style="dark")


plt.figure(figsize=(10,7))            # figure scale(size) ----------------


sns.barplot(x="day",y="total_bill",data=d,hue="day", palette="CMRmap")

# plt.grid()                   # Matplotlib grid

sns.despine()             # sns remove axis from plot 

plt.show()
```

#### Context ----------------------

```python
sns.set_style("white")           # Seaborn grid
# sns.set(style="dark")

sns.set_context(context="paper",font_scale=1.5)             # "paper", "notebook", "talk", "poster" ----------
 
sns.barplot(x="day",y="total_bill",data=d,hue="day", palette="cool")

# plt.grid()                   # Matplotlib grid

plt.show()
```
