### Pair Plot -------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
D=sns.load_dataset("penguins")
D
```

```python
sns.pairplot(D)
plt.show()
```

```python
D=sns.load_dataset("tips")
D
```

```python
sns.pairplot(D)
plt.show()
```

```python
sns.pairplot(D,hue="sex")
plt.show()
```

```python
select_column=["total_bill","tip"]
sns.pairplot(D[select_column+["sex"]],hue="sex", hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(D,hue="sex",palette="Accent",kind="kde")  # scatter, hist, kde, reg
plt.show()
```

```python
sns.pairplot(D,hue="sex",kind="kde",diag_kind="hist")
plt.show()
```

```python
sns.pairplot(D,hue="sex",markers=["*","<"])
plt.show()
```
