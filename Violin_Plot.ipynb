### Violin Plot -----------------------

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
D=sns.load_dataset("tips")
D
```

```python
sns.violinplot(x="day",y="total_bill",data=D,color="r")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=D,hue="time",linewidth=1,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=D,hue="day", palette="Dark2",order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=D,hue="sex",split=True,inner="stick")
plt.show()
```

```python
sns.violinplot(x=D["sex"]) # y=D["sex"]
plt.show()
```
